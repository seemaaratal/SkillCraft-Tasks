```python
import pandas as pd

df = pd.read_excel(
    r"C:\Users\seema\Downloads\Sample-Superstore.xlsx",
    engine="openpyxl"
)

df.head()
```

```python
# Check first 5 rows
df.head()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Check duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
# Convert dates
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Check data types
df.dtypes
```

```python
df.to_csv("Cleaned_Superstore.csv", index=False)
```
